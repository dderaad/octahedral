```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sim
```

```python
seed = 54421
rng = np.random.default_rng(seed)
N = 1_000
results, XYZ = sim.sim(seed, N, save_edges=True, save_points=True)

np.mean(results["regular"])
```

```python
results
```

```python
is_regular = results["regular"].astype(bool)
regs = results.loc[is_regular].head(3)
irregs = results.loc[~is_regular].head(3)
examples = pd.concat([regs, irregs])
display(examples)
display(examples.index)
display(XYZ[examples.index, :])
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

edge_traces = []
vertex_traces = []

for idx, row in examples.iterrows():
    xyz = XYZ[idx, :].reshape(6, 3)
    vertex_trace = go.Scatter3d(x = xyz[:, 0], y = xyz[:, 1], z = xyz[:, 2], text=[f"regular={is_regular[idx]}"]*6, mode="markers")
    xe, ye, ze = [], [], []
    for i, j in row[1:]:
        xe += [xyz[i, 0], xyz[j, 0], None]
        ye += [xyz[i, 1], xyz[j, 1], None]
        ze += [xyz[i, 2], xyz[j, 2], None]

    edge_trace = go.Scatter3d(
        x=xe,
        y=ye,
        z=ze,
        mode="lines",
        line=dict(width=4),
        name="Edges",
    )

    edge_traces.append(edge_trace)
    vertex_traces.append(vertex_trace)

fig = make_subplots(
    rows=3,
    cols=2,
    specs=[[{"type": "scene"}] * 2] * 3, 
    subplot_titles=["regular = True", "regular = False"] + 2 * 2 * [""],
    horizontal_spacing=0,
    vertical_spacing=0,
    column_widths=[1, 1],
    row_heights=[1]*3,
    
)

fig.update_layout(
    showlegend=False,
    autosize=False,
    width=1400,
    height=1000
)
fig.update_scenes(
    xaxis = dict(visible=False),
    yaxis = dict(visible=False),
    zaxis = dict(visible=False)
)

fig.print_grid()

for i, traces in enumerate(zip(edge_traces, vertex_traces)):
    fig.add_trace(traces[0], row=1 + i % 3, col=1 + i // 3)
    fig.add_trace(traces[1], row=1 + i % 3, col=1 + i // 3)


fig.show()
```
